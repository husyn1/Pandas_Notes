## Indexes 

```python
import pandas as pd
```

```python
customers = {
    "firstname":["Sam","Tom","Phil"],
    "lastname":["johnson","Denis","Alli"],
    "email":["SamJ123@gmail.com","Tom_Den65@yahoo.com","Phil_Alli202@gmail.com"]
}


customers_df=pd.DataFrame(customers)
customers_df
#  the first column with numbers and no column name is called the index
```

```python
customers_df.email
```

```python
#  change the index
customers_df.set_index("email",inplace=True) 
# inplace = False is used to display the change but does not change the actual dataframe
```

```python
customers_df
```

```python
# if u want to reset your index back to the default
customers_df.reset_index(inplace=True)
```

```python
customers_df
```
